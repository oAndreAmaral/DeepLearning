```python
from skimage import io
import numpy as np
import pandas as pd
import cv2 as cv
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Import Image

```python
fig1 = io.imread('Img1.tif')
fig2 = io.imread('Img2.tif')
fig3 = io.imread('Img3.tif')
fig4 = io.imread('Img4.tif')
fig5 = io.imread('Img5.tif')
fig6 = io.imread('Img6.tif')
fig7 = io.imread('Img7.tif')
```

```python
io.imshow(fig1)
```

# Convert Image To GrayScale

```python
src1 = cv.cvtColor(fig1, cv.COLOR_RGB2GRAY)
src2 = cv.cvtColor(fig2, cv.COLOR_RGB2GRAY)
src3 = cv.cvtColor(fig3, cv.COLOR_RGB2GRAY)
src4 = cv.cvtColor(fig4, cv.COLOR_RGB2GRAY)
src5 = cv.cvtColor(fig5, cv.COLOR_RGB2GRAY)
src6 = cv.cvtColor(fig6, cv.COLOR_RGB2GRAY)
src7 = cv.cvtColor(fig7, cv.COLOR_RGB2GRAY)
```

```python
io.imshow(src1)
```

# Histogram Equalization

```python
dst1 = cv.equalizeHist(src1)
dst2 = cv.equalizeHist(src2)
dst3 = cv.equalizeHist(src3)
dst4 = cv.equalizeHist(src4)
dst5 = cv.equalizeHist(src5)
dst6 = cv.equalizeHist(src6)
dst7 = cv.equalizeHist(src7)
```

# GLCM Matrix

```python
from skimage.feature import greycomatrix, greycoprops

glcm1 = greycomatrix(dst1, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
glcm2 = greycomatrix(dst2, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
glcm3 = greycomatrix(dst3, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
glcm4 = greycomatrix(dst4, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
glcm5 = greycomatrix(dst5, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
glcm6 = greycomatrix(dst6, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
glcm7 = greycomatrix(dst7, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
```

# Textures Features Extraction

```python
dissimilarity1 = []
dissimilarity2 = []
dissimilarity3 = []
dissimilarity4 = []
dissimilarity5 = []
dissimilarity6 = []
dissimilarity7 = []

dissimilarity1.append(greycoprops(glcm1, 'dissimilarity')[0, 0])
dissimilarity2.append(greycoprops(glcm2, 'dissimilarity')[0, 0])
dissimilarity3.append(greycoprops(glcm3, 'dissimilarity')[0, 0])
dissimilarity4.append(greycoprops(glcm4, 'dissimilarity')[0, 0])
dissimilarity5.append(greycoprops(glcm5, 'dissimilarity')[0, 0])
dissimilarity6.append(greycoprops(glcm6, 'dissimilarity')[0, 0])
dissimilarity7.append(greycoprops(glcm7, 'dissimilarity')[0, 0])
```

#### Entropia

```python
from scipy.stats import entropy

log_m1 = np.zeros((256,256))
log_m2 = np.zeros((256,256))
log_m3 = np.zeros((256,256))
log_m4 = np.zeros((256,256))
log_m5 = np.zeros((256,256))
log_m6 = np.zeros((256,256))
log_m7 = np.zeros((256,256))

# --> 1
for i in range(0,256):
    for j in range(0,256):
        if (glcm1[i,j] == 0):
            log_m1[i,j] = 0
        if (glcm1[i,j] != 0):
            log_m1[i,j] = np.log(glcm1[i,j])

# --> 2
for i in range(0,256):
    for j in range(0,256):
        if (glcm2[i,j] == 0):
            log_m2[i,j] = 0
        if (glcm2[i,j] != 0):
            log_m2[i,j] = np.log(glcm2[i,j])

# --> 3
for i in range(0,256):
    for j in range(0,256):
        if (glcm3[i,j] == 0):
            log_m3[i,j] = 0
        if (glcm3[i,j] != 0):
            log_m3[i,j] = np.log(glcm3[i,j])
            
# --> 4
for i in range(0,256):
    for j in range(0,256):
        if (glcm4[i,j] == 0):
            log_m4[i,j] = 0
        if (glcm4[i,j] != 0):
            log_m4[i,j] = np.log(glcm4[i,j])
            
# --> 5
for i in range(0,256):
    for j in range(0,256):
        if (glcm5[i,j] == 0):
            log_m5[i,j] = 0
        if (glcm5[i,j] != 0):
            log_m5[i,j] = np.log(glcm5[i,j])
            
# --> 6
for i in range(0,256):
    for j in range(0,256):
        if (glcm6[i,j] == 0):
            log_m6[i,j] = 0
        if (glcm6[i,j] != 0):
            log_m6[i,j] = np.log(glcm6[i,j])
            
# --> 7
for i in range(0,256):
    for j in range(0,256):
        if (glcm7[i,j] == 0):
            log_m7[i,j] = 0
        if (glcm7[i,j] != 0):
            log_m7[i,j] = np.log(glcm7[i,j])
```

```python
entropy1 = -np.sum(np.multiply(log_m1, glcm1[:,:,0,0]))
entropy2 = -np.sum(np.multiply(log_m2, glcm2[:,:,0,0]))
entropy3 = -np.sum(np.multiply(log_m3, glcm3[:,:,0,0]))
entropy4 = -np.sum(np.multiply(log_m4, glcm4[:,:,0,0]))
entropy5 = -np.sum(np.multiply(log_m5, glcm5[:,:,0,0]))
entropy6 = -np.sum(np.multiply(log_m6, glcm6[:,:,0,0]))
entropy7 = -np.sum(np.multiply(log_m7, glcm7[:,:,0,0]))
```
